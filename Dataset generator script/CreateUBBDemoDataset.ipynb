```python
# used to randomly select a word from a sample

def chooseFrom(set):
  return set[rand.randint(0,len(set)-1)]
```

```python
# used to add two arrays in one array, to make it easier to write the corresponding label to each word/pair of words

def addPair( concept, labels):
  array = []
  # check if this is really needed
  if len(concept) == 1 and len(labels) == 1:
    array.append([concept, labels])
  else:
    for i, j in zip(concept, labels):
      array.append([i,j])
  return array
```

```python
#function to split each sample where a word is either separated by - or by space

def splitWords(array, sep):
  return [re.split(sep,x) for x in array]
```

```python
# function to write the file where each word is written on the same line with its label, separated by space
# each sentence is separated by space one from another

def writeFile(array, filename):
  f = open(filename, "a")
  for i,k in enumerate(array):
    # if there is only one word, then write the associated B-XXX label
      if len(array[i][0]) == 1:
        f.write(str(array[i][0][0]) + " -X- -X- " + str(array[i][1][0]) + "\n")
      else:
        # check for None items, generated by the split method
        if array[i][0][0] is not None:
          # attach the B-XXX label to the first word and I-XXX to the rest
          f.write(array[i][0][0] + " -X- -X- " + array[i][1][0] + "\n")
          for j in array[i][0][1:]:
            if j is not None:
              f.write(j + " -X- -X- " + str(array[i][1][1]) + "\n")
  f.write("\n")
  f.close()
```

```python
# read a file in order to create the samples

def readFile(filename):
  f = open(filename,"r")
  ft = f.read().splitlines()
  f.close()
  return ft
```

```python
# used when two files need to be merged in one array, by concatenating their lines 

def appendFiles(filename1,filename2):
  L = []
  f = open(filename1,"r")
  e = open(filename2,"r")
  ft = e.read().splitlines()
  for line in f:
    L.append(line.rstrip() + (ft[rand.randint(0,6)]))
  f.close() 
  e.close()
  return L
```
